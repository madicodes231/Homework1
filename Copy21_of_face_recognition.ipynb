# Passing images to train

```
import face_recognition
import numpy as np
from PIL import Image, ImageDraw
from IPython.display import display


# face_1 = face_recognition.load_image_file("elon.jpg")
# face_1_encoding = face_recognition.face_encodings(face_1)[0]

face_1 = face_recognition.load_image_file("Aaron Peirsol.jpg")
face_1_encoding = face_recognition.face_encodings(face_1)[0]

face_2 = face_recognition.load_image_file("Adam Sandler.jpg")
face_2_encoding = face_recognition.face_encodings(face_2)[0]

face_3 = face_recognition.load_image_file("Adam Scott.jpg")
face_3_encoding = face_recognition.face_encodings(face_3)[0]

face_4 = face_recognition.load_image_file("Ben Affleck.jpg")
face_4_encoding = face_recognition.face_encodings(face_4)[0]

face_5 = face_recognition.load_image_file("Donald Trump.jpg")
face_5_encoding = face_recognition.face_encodings(face_5)[0]

face_6 = face_recognition.load_image_file("Mike Pence.jpg")
face_6_encoding = face_recognition.face_encodings(face_6)[0]

face_7 = face_recognition.load_image_file("zy.jpg")
face_7_encoding = face_recognition.face_encodings(face_7)[0]

known_face_encodings = [
    face_1_encoding,
    face_2_encoding,
    face_3_encoding,
    face_4_encoding,
    face_5_encoding,
    face_6_encoding,
    face_7_encoding
]
known_face_names = [
    "Aaron Peirsol",
    "Adam Sandler","Adam Scott",
    "Ben Affleck","Donald Trump",
    "Mike Pence", "Zydrunas Ilgauskas"
]
print("Done learning and creating profiles")
```

# Testing Face Recognition using Images

```
unknown_image1 = face_recognition.load_image_file("Aaron_Peirsol_0003.jpg")
unknown_image2 = face_recognition.load_image_file("Adam_Sandler_0004.jpg")
unknown_image3 = face_recognition.load_image_file("Adam_Scott_0002.jpg")
unknown_image4 = face_recognition.load_image_file("Ben_Affleck_0007.jpg")
unknown_image5 = face_recognition.load_image_file("Zydrunas_Ilgauskas_0001.jpg")
unknown_image6 = face_recognition.load_image_file("OIP (3).jpg")
unknown_image7 = face_recognition.load_image_file("donald.jpg")

face_locations = face_recognition.face_locations(unknown_image2)
face_encodings = face_recognition.face_encodings(unknown_image2, face_locations)

pil_image = Image.fromarray(unknown_image2)
draw = ImageDraw.Draw(pil_image)

for (top, right, bottom, left), face_encoding in zip(face_locations, face_encodings):
    # See if the face is a match for the known face(s)
    matches = face_recognition.compare_faces(known_face_encodings, face_encoding)

    name = "Unknown"

    face_distances = face_recognition.face_distance(known_face_encodings, face_encoding)
    best_match_index = np.argmin(face_distances)
    if matches[best_match_index]:
        name = known_face_names[best_match_index]

    # Draw a box around the face using the Pillow module
    draw.rectangle(((left, top), (right, bottom)), outline=(0, 255, 255))
    print(name)

    # Draw a label with a name below the face
    text_width, text_height = draw.textsize(name)
    # draw.rectangle(((left, bottom - text_height - 10), (right, bottom)), fill=(0, 0, 255), outline=(0, 0, 255))
    draw.text((left + 6, bottom - text_height + 25), name, fill=(255, 255, 255, 255))


# Remove the drawing library from memory as per the Pillow docs
del draw

# Display the resulting image
display(pil_image)
```
