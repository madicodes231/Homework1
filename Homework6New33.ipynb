```python
!pip install requests

import requests

target_url ='http://www.gutenberg.org/files/2265/2265.txt'

response = requests.get(target_url)

data = response.text
```

```python
from pathlib import Path

from textblob import TextBlob

blob=TextBlob(data)
```

```python
from nltk.corpus import stopwords

stop_words = stopwords.words('english')
```

```python
items = blob.word_counts.items()
```

```python
items = [item for item in items if item[0] not in stop_words]
```

```python
from operator import itemgetter
sorted_items = sorted(items, key=itemgetter(1), reverse=True)
```

```python
top20 = sorted_items[0:20]
```

```python
import pandas as pd
df = pd.DataFrame(top20, columns=['word', 'count'])
df
```

```python
axes = df.plot.bar(x='word', y='count', legend=False)
import matplotlib.pyplot as plt
plt.gcf().tight_layout()
```

```python
import imageio

image_file="https://media.cheggcdn.com/media/216/21621ee5-e80f-47f3-9145-513f2229b390/phploeBuh.png"

mask_image = imageio.imread(image_file)
```

```python
from wordcloud import WordCloud
wordcloud = WordCloud(width=1000, height=1000,
    colormap='prism', mask=mask_image, background_color='white')
```

```python
wordcloud = wordcloud.generate(data)
```

```python
wordcloud = wordcloud.to_file('HamletCloud.png')
```

```python
plt.figure(figsize=(10, 10), facecolor=None)
plt.axis("off")
plt.imshow(wordcloud)
```
